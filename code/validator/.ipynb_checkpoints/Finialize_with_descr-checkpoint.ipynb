## Company

```python
import pandas as pd
dff= pd.read_csv('Final_SnP500_companies_handles_2010to2020.csv')
dff
```

```python
dff.iloc[0]
```

```python
gvkeys=[]
companies=[]
names=[]
ids=[]
handles=[]
verified=[]
validated=[]
created_ats=[]
statuses_counts=[]
followers_counts=[]
descriptions=[]


for i in range(len(dff)):
    j= dff.iloc[i]['Handle']
    try:
        obj=api.get_user(j)
        gvkeys.append(dff.iloc[i]['GVKEY'])
        companies.append(dff.iloc[i]['Company'])
        names.append(obj.name)
        ids.append(obj.id)
        handles.append(j)
        verified.append(obj.verified)
        validated.append(dff.iloc[i]['validated'])
        created_ats.append(obj.created_at)
        statuses_counts.append(obj.statuses_count)
        followers_counts.append(obj.followers_count)
        descriptions.append(obj.description)
    except:
        print('bad user name,', j)
        pass

columns = ['GVKEY','Company', 'TwitterID', 'TwitterName', 'TwitterHandle', 'Verified(Y/N)', 'Validated', 'Since', 'TweetsCount', 'FollowersCount', 'TwitterDescr']
df = pd.DataFrame(list(zip(gvkeys,companies,ids,names,handles,verified,validated,created_ats,statuses_counts,followers_counts,descriptions)), columns =columns) 
df.to_csv("Final_SnP500_companies_handles_2010To2020.csv", index=False)
```
