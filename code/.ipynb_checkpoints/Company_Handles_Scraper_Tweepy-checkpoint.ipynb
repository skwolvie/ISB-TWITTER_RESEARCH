```python
df['Company']= df['Company'].str.replace(r"\(.*\)","")
df['Company']= df['Company'].str.replace(' -CL B', '')
df['Company']= df['Company'].str.replace('/DE', '')
df['Company']= df['Company'].str.replace('/ME', '')
df['Company']= df['Company'].str.replace('ARCHER-DANIELS-MIDLAND CO', 'ARCHER-DANIELS-MIDLAND')
df['Company']= df['Company'].str.strip()
df
```

```python
tdf=pd.DataFrame()
companies=[]
handles=[]
descriptions=[]
verified=[]
badnames=[]
for i in range(len(df)):
    users = api.search_users(df.iloc[i]['Company'])                     
    lst=[]
    for user in users:
        lst.append(user.screen_name)
    if len(lst)==0:
        badnames.append(df.iloc[i]['Company'])
    for j in lst:
        result=api.get_user(j).verified
        if result==True:
            companies.append(df.iloc[i]['Company'])
            handles.append(j)
            descriptions.append(api.get_user(j).description)
            verified.append(result)
            print(df.iloc[i]['Company'], j)
```

```python
columns = ['Company', 'Handle', 'Twitter_Descr', 'verified']
tdf = pd.DataFrame(list(zip(companies, handles, descriptions, verified)), columns =columns) 
tdf.dropna(inplace=True)
tdf= tdf[tdf['verified']!=False]
tdf.info()
```
