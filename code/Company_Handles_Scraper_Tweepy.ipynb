```python
from collections import Counter
cnt = Counter()
for text in df["Company"].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(30)
```

```python
df['Company']= df['Company'].str.replace(' INC', '')
df['Company']= df['Company'].str.replace(' CORP', '')
df['Company']= df['Company'].str.replace(' CO', '')
df['Company']= df['Company'].str.replace(' GROUP', '')
df['Company']= df['Company'].str.replace(' GRP', '')
df['Company']= df['Company'].str.replace(' LTD', '')
df['Company']= df['Company'].str.replace(' BRANDS', '')
df['Company']= df['Company'].str.replace(' -CL', '')
```

```python
tdf=pd.DataFrame()
companies=[]
handles=[]
descriptions=[]
verified=[]

for i in range(len(df)):
    users = api.search_users(df.iloc[i]['Company'])                
        
    lst=[]
    for user in users:
        lst.append(user.screen_name)
    for j in lst:
        result=api.get_user(j).verified
        if result==True:
            companies.append(df.iloc[i]['Company'])
            handles.append(j)
            descriptions.append(api.get_user(j).description)
            verified.append(result)
            print(df.iloc[i]['Company'], j)
```

```python
columns = ['Company', 'Handle', 'Twitter_Descr', 'verified']
tdf = pd.DataFrame(list(zip(companies, handles, descriptions, verified)), columns =columns) 
tdf.dropna(inplace=True)
tdf= tdf[tdf['verified']!=False]
tdf.info()
```

```python
tdf.to_csv("../output/Company_handles_tweepy_lookup.csv", index=False)
```
