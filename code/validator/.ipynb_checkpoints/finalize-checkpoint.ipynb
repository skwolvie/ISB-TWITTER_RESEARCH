```python
import pandas as pd
import numpy as np
import os
```

```python
os.listdir()
```

```python
df= pd.read_excel('badnames.xlsx')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['validated']=1
df
```

```python
df1= pd.read_excel('singlehandles.xlsx')
df1.info()
```

```python
df1.validated.value_counts()
```

```python
df1=df1[df1.validated==1]
df1.shape
```

```python
df1.drop_duplicates(inplace=True)
df1.shape
```

```python
df2= pd.read_excel('multiplehandles.xlsx')
df2.info()
```

```python
df2.dropna(inplace=True)
df2.info()
```

```python
df2.Validation.value_counts()
```

```python
df2=df2[df2.Validation==1]
df2.shape
```

```python
print(df.columns)
print(df1.columns)
print(df2.columns)
```

```python
df2.rename(columns={'Validation':'validated'}, inplace=True)
df2
```

```python
dff= pd.concat([df, df1, df2], axis=0)
dff
```

```python
dff.drop_duplicates(inplace=True)
dff
```

```python
dff.info()
```

```python
dff.to_csv('Final_SnP500_companies_handles_2010to2020.csv', index=False)
```
